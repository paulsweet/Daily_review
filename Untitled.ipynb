```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
x_count = np.linspace(0,100000,1000000)
q2 = 23520-0.5*x_count
q21 = 2*24570-2*x_count 
# point=[[17080,14980]]

# x1_tmp = np.linspace(0,17080,100)
# y1_tmp = np.linspace(0,14980,100)


fig,ax = plt.subplots(figsize=(10,10))
ax.plot(x_count,q2,c='g')
ax.plot(x_count,q21,c='g')
# sns.lineplot(x=x1_tmp,y=14979,c='r')
# sns.lineplot(x=4,y=y1_tmp,c='r')


ax.scatter([17080],[14980],c="g", marker="o")
ax.set_xlim([0, 60000])
ax.set_ylim([0,60000])
ax.set_xticks([0,17080,24570,47040])
ax.set_yticks([14980,23520,2*24570])
ax.set_xlabel(r'${ Q }_{ 1 }$',fontsize=15)
ax.set_ylabel(r'${ Q }_{ 2 }$',fontsize=15)
ax.text(20000, 15000,  r"Nash equilibrium",color='black',fontsize=15)
ax.text(10000,40000,Sr"${ Q }_{ 1 }=24570-0.5{ Q }_{ 2 }$",color= 'black',fontsize=15)
ax.text(43000,6000,r"${ Q }_{ 2 }=23520-0.5{ Q }_{ 1 }$",color= 'black',fontsize=15)

```
