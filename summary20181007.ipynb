## 2.plt和面对对象的方式画图

Matplotlib对象简介

### 2.1 创建子图(注意和多图的区别)

````python
import matplotlib.pyplot as plt
import numpy as np
#建立一个fig对象，建立一个axis对象,这里等价于：
fig,ax = plt.subplots(2,3,figsize=(15,10)) # 两行，三列
# ax[*,*]的形式来写每一张图
ax[0,0].grid(linewidth='2')
````

````python
fig = plt.figure(figsize=(15,5)) #figsize这里的参数，第一个15是横向长度，第二个是纵向，第三个是图的数量，对象是整个figure

ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)
````

### 2.2 加网格

````python
fig = plt.figure(figsize=(15,5)) #figsize这里的参数，第一个15是横向长度，第二个是纵向，第三个是图的数量，对象是整个figure

ax1 = fig.add_subplot(131)
ax1.grid(True)

ax2 = fig.add_subplot(132)
ax2.grid(linestyle='--',color='g')

ax3 = fig.add_subplot(133)
ax3.grid(linewidth='1.5',color='r',linestyle='-')
````
